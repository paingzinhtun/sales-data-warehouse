# Product Performance Analysis

## Objective
Analyze the performance of products based on sales, revenue, and trends.

## Steps:
1. Load cleaned sales data.
2. Aggregate total sales per product.
3. Identify top-performing products.
4. Visualize trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Database connection
DATABASE_URL = "postgresql://postgres:8484123@localhost:5432/sales_db"
engine = create_engine(DATABASE_URL)
```

```python
# Load cleaned data
query = "SELECT product_id, SUM(total_sales) as total_revenue FROM fact_sales GROUP BY product_id ORDER BY total_revenue DESC"
df = pd.read_sql(query, engine)
```

```python
# Visualize top products
plt.figure(figsize=(10,5))
plt.bar(df['product_id'][:10], df['total_revenue'][:10], color='skyblue')
plt.xlabel("Product ID")
plt.ylabel("Total Revenue")
plt.title("Top 10 Best-Selling Products")
plt.show()
```
